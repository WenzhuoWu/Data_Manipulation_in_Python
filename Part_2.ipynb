__Exercise 1.1 (10 points).__ Give three examples of indexing a data frame with `[ ]`, `.loc[ ]`, and `.iloc[ ]`, respectively. Explain how each of these indexing methods is different.

```python
import pandas as pd
df = pd.DataFrame({"x": [1, 2, 3], "y": ["a", "b", "c"]}) #generate a data frame first
df
```

```python
display(df["x"])
display(df.x[1])
df [['x','y'] ]  
```

**[ ] can be used to select object by position, name, or condition.**

**First example shows we select the column which name is "x".**

**Second example shows we select the second object of column x.**

**Third example shows we select two columns which are x and y.**

```python
display(df.loc[:, "x"])
display(df.loc[df.x ==1])
display(df.loc[(df.x > 0) &(df.y!= 'c')])
```

**.loc[ ] can be used to select object by name or condition.**

**First example shows we select all the rows for column x.**

**Second example shows we select the row when x is 1.**

**Third example shows we select the row when x is greater than 0 and y not equal to c.**

```python
display(df.iloc[1,1])
display(df.iloc[1:3,1])
display(df.iloc[:,1])
```

**.iloc[ ] can be used to select object by position.**

**First example shows we select the object at second row and second column.**

**Second example shows we select the second and third row (index is 1 and 2) at the second column (y).**

**Third example shows we select all the rows at the second column (y).**

**In this example, we have to set the index as Name, otherwise it will align the data frame based on the default index. Also, the number of rows should be the same, otherwise it will only align the rows with same index.**

```python
x = pd.Series([10, 20, 30], index = ["a", "b", "c"])
y = pd.Series([10, 20, 30, 10], index = ["b", "a", "d", "c"])
display(x)
display(y)
x+y
```

**In this example, elements are automatically aligned by index. For index a, 10+20=30. They are not alighned by position.**

**The feature of Pandas' data alignment is that it can merge the two dataframes based on the index. It's a powerful tool when you have multiple variables in different dataset measured from same group of objects. It can merge them automaticlly. But if we want to have operation based on position, we can reset the index to add another column as index. Then the data can align based on position and order.**

```python
list= df_5.columns[3:].tolist() # choose the column which contain dates
df_6= pd.melt(df_5, id_vars = ['Admin2'], value_vars= list,var_name='Date', value_name='Death')
df_6
```

```python
df_6['Date'] = pd.to_datetime(df_6['Date']) #change Date type to datetime
# https://pandas.pydata.org/docs/reference/api/pandas.PeriodIndex.html.
```

```python
import datetime #https://www.tutorialspoint.com/How-can-I-get-the-current-week-using-Python#:~:text=YOu%20can%20use%20the%20isocalender,number%20and%20Day%20of%20Week.
#write a function to generate week information from a date
def week_func(my_date):
    my_date = pd.to_datetime(my_date) #change type
    year, week_num, day_of_week = my_date.isocalendar() #extract week number and year
    i= ("Week #" + str(week_num) + " of year " + str(year))
    return (i)
```

```python
df_6['Week'] = df_6['Date'].apply(week_func) #apply week_func on column 'date' 
display(df_6.tail())
df_7 = df_6.groupby('Week').sum() #group the data and sum them based on Week
display(df_7)
df_8= df_6.groupby('Week').mean() #group the data and calculate the mean based on Week
df_8.idxmax(axis = 0) #find the maximum 
```
